```python
# http://www.fhfa.gov/DataTools/Downloads/Pages/Public-Use-Databases.aspx
pd.set_option('max_rows',10,'max_columns',16)
```

```python
# read in the data as text
df = pd.concat([ pd.read_csv(
    "{0}_sf2013a_loans.txt".format(f),
    sep='\s+',header=None,index_col=None) for f in ['fhlmc','fnma']])

df.columns = ['vendor','record_number','msa','per_minority',
              'tract_income_ratio','borrow_income_ratio','ltv','purpose','guarantee','borrow_ethnicity',
              'coborrow_ethnicity','borrow_gender','coborrow_gender','unit','affordability']
            
# write to hdf5
df.to_hdf('housing.h5','df',mode='w')
df.info(null_counts=True)
```

```python
df = pd.read_hdf('housing.h5','df')
df.head()
```

```python
# assign meaningful values
df['vendor'] = pd.Categorical(df['vendor'],ordered=False).rename_categories(['fannie','freddie'])
df['msa'] = pd.Categorical(df['msa'],ordered=False).rename_categories(['non_metro','metro'])

df['per_minority'] = df['per_minority'].astype('category').cat.rename_categories(['0-10','10-30','30-100',np.nan])
df['tract_income_ratio'] = df['tract_income_ratio'].astype('category').cat.rename_categories(['0-80','80-120','120+',np.nan])

df['borrow_income_ratio'] = df['borrow_income_ratio'].astype('category').cat.rename_categories(['0-50','50-80','80+',np.nan])
df['ltv'] = df['ltv'].astype('category').cat.rename_categories(['0-60','60-80','80-90','90-95','95+',np.nan])

df['purpose'] = pd.Categorical(df['purpose'],ordered=False).rename_categories(['purchase','other',np.nan])
df['guarantee'] = pd.Categorical(df['guarantee'],ordered=False).rename_categories(['FHA/VA','RHS','Conv'])

df['affordability'] = pd.Categorical(df['affordability'],ordered=False).rename_categories([np.nan,'LI','VLI LIA','VLI not LIA','Other'])
```

```python
df['vltv'] = df['ltv'].cat.rename_categories(range(6)).astype(int)
df['vltv'].describe()
```

```python
df.groupby(['per_minority','borrow_income_ratio'])['vltv'].mean().order()
```

```python
g = ['per_minority','borrow_income_ratio']
z = df.dropna(subset=g).groupby(g)['ltv'].value_counts()
z = z.reset_index()
z.columns=list(z.columns[0:2]) + ['ltv','count']
z = z.dropna(subset=['ltv'])
z
```

```python
fig, axes = plt.subplots(3, 2, figsize=(10, 10));
plt.subplots_adjust(wspace=0.5, hspace=0.5);
grouper = z.groupby('ltv')
for ax, (name, grp) in zip(axes.ravel()[0:len(grouper)],grouper):
    grp.set_index(g).plot(kind='bar',ax=ax,subplots=True,legend=False)
```

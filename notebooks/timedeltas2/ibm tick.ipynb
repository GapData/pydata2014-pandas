ibm tick data from 10/30/14
- showing tz conversions
- resample & freq counting

```python
pd.set_option('max_rows',14)

def plot(obj,**kwargs):
    obj = obj.copy()
    obj.index = obj.index.format()
    obj.plot(**kwargs)
```

```python
def get_data():
    import urllib as web
    date = "2014-10-30"

    stock='IBM.N'
    baseUrl='http://netfonds.no/quotes/tradedump.php?csv_format=csv'
    fullUrl=baseUrl+'&date='+date.replace("-","")+'&paper='+stock

    webBuffer=web.urlopen(fullUrl)
    webData=pd.read_csv(webBuffer)
    webData.to_csv('ibm.csv')

# read in the data as a csv
df = pd.read_csv('ibm_tick_20141030.csv',
                 index_col=0, usecols=['time','price','quantity'])

# convert the index from CET to US tz, and drop it
df.index = pd.to_datetime(df.index).tz_localize('CET').tz_convert('US/Eastern').tz_localize(None)
df.info()
```

```python
df['price'].resample('1T',how='mean').plot()
```

```python
# create a TimedeltaIndex
df2 = df.copy()
df2.index = df.index-df.index[0]
df2.index
```

```python
# trade frequency
df.index.value_counts().sort_index().plot()
```

```python
# trade frequency
plot(df2.index.value_counts().sort_index())
```

```python
df['price'].resample('1T',how='count').plot()
```

```python
df2.resample('10T',how='ohlc')
```

```python
resampled = df2.resample('10T',how='ohlc')
closes = resampled.loc[:,('quantity','close')]
closes
```

```python
closes.order(ascending=False)
```

```python
plot(closes)
```

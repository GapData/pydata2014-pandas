```python
# read in the data as text
df = pd.read_csv('IBM_adjusted.txt',header=None)
df.columns = ['date','time','open','high','low','close','volume']

# combine the date/time as a string, convert to datetime64[ns]
df['date'] = pd.to_datetime(df['date'] + ' ' + df['time'],infer_datetime_format=True)
df.pop('time')

# set the date as the index
df = df.set_index('date')

# write to hdf5
df.to_hdf('ibm.h5','df',mode='w')
```

```python
# create the time differences
diffs = df.index.to_series().diff()
diffs.head()
```

```python
# per return of each period relative to each day in bp
g = df.groupby(pd.Grouper(level='date',freq='D'))

# take the first element of each group and propogate forward by date
closes = g['close'].first().reindex(df.index,method='pad')
df['rclose'] = 100*(df['close']-closes)/closes
```

```python
df
```

```python
# create relative times
rc = df[['rclose']].copy()
rc['days'] = g.first().index.to_series().reindex(df.index,method='pad')
rc['diff'] = rc.index-rc['days']
rc
```

```python
rc.pop('days')
rc = rc.reset_index().set_index('diff')
rc
```

```python
rc.index
```

```python
indexer = rc.date>=Timestamp('20141001')
df.loc[indexer.values,['close','rclose']].plot(subplots=True,figsize=(10,10))
oct = rc[indexer]
oct.describe()
```

```python
# plot a 10 minutes resample with ohlc
result = rc.loc[indexer,'rclose'].resample('10min',how='ohlc')
result.index = result.index.format()
result.plot(subplots=True,figsize=(10,10),layout=(2,2))
```
